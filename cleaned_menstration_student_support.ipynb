### Clean the Data 

```python
#both csvs have same same questions so merge vertically
menst_data = pd.concat([random_selection, online_selection], axis=0, ignore_index=True)
#matches the total data collection (so correct merge)
menst_data
```

```python
# List of columns to be deleted
columns_to_delete = [
                     'What other locations would like to see have free menstrual products?','Were you able to access your preferred type of product?  ',
       'Which type of menstrual products would you like to have access to?']


# Delete the specified columns
menst_data1 = menst_data.drop(columns=columns_to_delete)

#After the question "Have you used the free menstrual products" if it is no then only questions about support is answered
menst_data1
```

```python
menst_data1.columns
```

```python
#rename columns concisely
menst_data1.rename(columns={
    "Are you a full time or part time student?": "attendance_status",
    "Is using menstrual products applicable to you?": "menstruation_applicable",
    "Have you used the free menstrual products?": "free_products_used",
    "If the school provides free menstrual products. Would you use the free products or your personal ones?": "product_use_preference",
    "How often do you access the free sanitary products currently provided in the bathrooms?   ": "product_usage_frequency",
    "Which of the following describes your reasons for accessing the sanitary products at school?": "school_usage_reasoning",
    "Have you ever missed class due to not having access to menstration products. ": "missed_class_impact",
    "Would free menstrual products increase your comfortability when attending school?": "attendance_increase_impact",
    "In what way(s) can the availability of free sanitary products at school will have a positive impact on you?  ": "free_product_impact",
    "How Supportive are you towards CCM's Initiative of Free Period Products?": "support_scale",
    "If you are not supportive of this initiative what are the reasons?": "no_support_reason"
}, inplace=True)
menst_data1
```
